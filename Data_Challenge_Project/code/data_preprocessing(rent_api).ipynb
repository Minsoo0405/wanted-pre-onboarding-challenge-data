```python
i = 1

while i < len(json_data):
    if json_data[i].keys() == {"rentData"}:
      df = pd.json_normalize(json_data[i]['rentData']['row'])
      json_df = pd.concat([json_df, df], axis=0)

      if len(df) != len(json_df):
        print(f"{i}Î≤àÏß∏ DataFrame concat ÏÑ±Í≥µ")

      i += 1
    else:
      i += 1

print("DataFrame concat ÏôÑÎ£å!!!")
```
