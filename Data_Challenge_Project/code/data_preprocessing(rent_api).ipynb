### 2.데이터 확인하기

#### 2.1 json 데이터 구조 확인

```python
print(json_data[0].keys())
print(json_data[0]['rentData'].keys())
print(len(json_data))
print(json_data[0]['rentData']['row'])
```

#### 2.2 json 데이터를 데이터프레임 형태로 확인

```python
json_df = pd.json_normalize(json_data[0]['rentData']['row'])
json_df.head()
```

### 3.2018년 1월 데이터 DataFrame으로 합치기

#### 3.1 하나의 데이터프레임으로 합치기

```python
i = 1

while i < len(json_data):
    df = pd.json_normalize(json_data[i]['rentData']['row'])
    json_df = pd.concat([json_df, df], axis=0)

    if len(df) != len(json_df):
        print(f"{i}번째 DataFrame concat 성공")

    i += 1

print("DataFrame concat 완료!!!")
```

#### 3.3 index 재설정

```python
json_df.reset_index(drop=True, inplace=True)
```

#### 3.4 2월 데이터 삭제

```python
df = json_df[json_df['RENT_DT'] <= "2018-02-01 00:00:00"]
```

#### 4.3 컬럼별 정리

##### 4.3.1 필요없는 컬럼 삭제
- "RENT_HOLD", "RTN_HOLD", "START_INDEX", "END_INDEX", "RNUM", "DUP"

```python
df_1801 = df.drop(columns=["RENT_HOLD", "RTN_HOLD",
              "START_INDEX", "END_INDEX",
              "RNUM", "DUP"]).copy()
```

##### 4.3.2 컬럼별 데이터 타입 변환
- INTEGER : "USE_MIN", "BIRTH_YEAR"
- FLOAT : "USE_DST"

```python
df_1801.info()
```

```python
# integer - "USE_MIN" = "RTN_DT" - "RENT_DT" 계산 후, 데이터 타입 변환
df_1801["USE_MIN"] = (df_1801["RTN_DT"] - df_1801["RENT_DT"]).astype("int")
```

```python
# integer - "BIRTH_YEAR" null 값은 9999로 변환 후, 데이터 타입 변환
df_1801["BIRTH_YEAR"] = df_1801["BIRTH_YEAR"].fillna("9999").astype("int")
```

```python
df_1801[["USE_MIN", "BIRTH_YEAR"]].info()
```

##### 4.4.1 컬럼 생성

```python
df_1801.columns
```

```python
df_1801["Date"] = df_1801["RENT_DT"].dt.date
df_1801["Hour"] = df_1801["RENT_DT"].dt.hour
```

```python
df_1801[['Date', 'Hour']]
```

##### 4.4.2 대여건수 컬럼 생성

```python
df_fin = df_1801.groupby(["Date","Hour"]).agg({"Hour":"count"})
df_fin
```

```python
df_fin.rename(columns={"Hour":"Rent Bike Count"}, inplace=True)
```

```python
df_fin.reset_index()
```

# 📌 2018년 부터 2023년 까지 대여건수 csv 파일로 만들기

### 1. 데이터 불러오기

```python
import numpy as np
import pandas as pd
```

#### 1.1 vscode에서 불러오기

#### 1.2 colab에서 불러오기

```python
def load_json(y1,y2,m1,m2):
    # google drive 마운트 실행
    from google.colab import drive
    drive.mount('/content/drive')

    # json 파일 읽기
    import json

    json_data = []
    with open(f"/content/drive/MyDrive/rental_history_data/20{y1,y2}/{y1,y2,m1,m2}.json", "r", encoding="utf-8") as f:
        for line in f:
            json_data.append(json.loads(line))
```

```python
load_json(1,8,0,1)
```

### 2. json 데이터를 DataFrame으로 만들기

#### 2.1 DataFrame으로 변환

```python
i = 1

while i < len(json_data):
    if json_data[i].keys() == {"rentData"}:
      df = pd.json_normalize(json_data[i]['rentData']['row'])
      json_df = pd.concat([json_df, df], axis=0)

      if len(df) != len(json_df):
        print(f"{i}번째 DataFrame concat 성공")

      i += 1
    else:
      i += 1

print("DataFrame concat 완료!!!")
```

#### 2.2 해당 월 데이터가 아닌 데이터를 삭제

```python
df = json_df[json_df['RENT_DT'] <= f"20{y1}{y2}-{m1}{m2}-01 00:00:00"]
```

### 3. 데이터 타입 변환하기

#### 3.1 대여일시 데이터 타입 변환

```python
df['RENT_DT'] = pd.to_datetime(df['RENT_DT'], format= '%Y-%m-%d %H:%M:%S')
```

#### 3.2 반납일시 데이터 타입 변환

```python
df['RTN_DT'] = pd.to_datetime(df['RTN_DT'], format='%Y-%m-%d %H:%M:%S')
```

### 4. 중복 데이터 삭제하기


#### 4.1 대여일시가 같은 데이터 삭제

```python
# 대여일시가 중복된 데이터 삭제
df.drop_duplicates(subset=['RENT_DT'], keep='first', inplace=True)
```

### 5. 대여건수 관련 컬럼 생성하기

#### 5.1 "Date", "Hour" 컬럼 생성

```python
df["Date"] = df["RENT_DT"].dt.date
df["Hour"] = df["RENT_DT"].dt.hour
```

#### 5.2 "Rent Bike Count" 컬럼 생성

```python
df_rent = df.groupby(["Date","Hour"]).agg({"Hour":"count"})

df_rent.rename(columns={"Hour":"Rent Bike Count"}, inplace=True)
```

### 6. 데이터 프레임 저장하기

```python
df_rent.reset_index(inplace=True)
```
