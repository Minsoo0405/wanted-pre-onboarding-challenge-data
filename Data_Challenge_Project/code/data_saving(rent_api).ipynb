```python
import json

def load_json(a, b, c):
  json_data = []
  with open(f"/content/drive/MyDrive/rental_history_data/{a}/{b}{c}.json", "r", encoding="utf-8") as f:
    for line in f:
      json_data.append(json.loads(line))
    return json_data
```

### 2. json ÌòïÌÉúÎ•º Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏúºÎ°ú Î≥ÄÌôòÌïòÍ∏∞

```python
def trans_df(x):
    if len(x) < 2:
        print("ÏûÖÎ†• Î¶¨Ïä§Ìä∏Ïóê ÏöîÏÜåÍ∞Ä Ï∂©Î∂ÑÌïòÏßÄ ÏïäÏäµÎãàÎã§.")
        return

    i = 1

    if "rentData" in x[i] and x[i]["rentData"] and "row" in x[i]["rentData"]:
        json_df = pd.json_normalize(x[0]['rentData']['row'])

        while i < len(x):
            if "rentData" in x[i] and x[i]["rentData"] and "row" in x[i]["rentData"]:
                df = pd.json_normalize(x[i]['rentData']['row'])
                json_df = pd.concat([json_df, df], axis=0)

                if len(df) != len(json_df):
                    print(f"{i}Î≤àÏß∏ DataFrame concat ÏÑ±Í≥µ")

            i += 1
        return json_df
    else:
        print("DataFrame concat ÏôÑÎ£å!!!")
```

```python
df = trans_df(json_data)
```

#### 3.1 Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ Î≥ÄÌôò

```python
df['RENT_DT'] = pd.to_datetime(df['RENT_DT'], format= '%Y-%m-%d %H:%M:%S')
df['RTN_DT'] = pd.to_datetime(df['RTN_DT'], format='%Y-%m-%d %H:%M:%S')
```

#### 3.2 ÎåÄÏó¨ÏùºÏãú Ï§ëÎ≥µ ÏÇ≠Ï†ú

```python
def del_dup(x):
  # ÎåÄÏó¨ÏùºÏãúÍ∞Ä Ï§ëÎ≥µÎêú Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú
  filters = x.drop_duplicates(subset=['RENT_DT'], keep='first')
  return filters
```

```python
df = del_dup(df)
df
```

#### 3.3 Ìï¥Îãπ ÏõîÏù¥ ÏïÑÎãå Îç∞Ïù¥ÌÑ∞ ÏÇ≠Ï†ú

```python
def del_month(a,x,y):
  filters = (a['RENT_DT'].dt.year == int(x)) & (a['RENT_DT'].dt.month == int(y))
  return a[filters]
```

```python
df = del_month(df,year,mm)
df
```

#### 3.4 "Date", "Hour", "Bike" Ïª¨Îüº ÏÉùÏÑ±

```python
def make_col(a):
  # "Date", "Hour" Ïª¨Îüº ÏÉùÏÑ±
  a["Date"] = a["RENT_DT"].dt.date
  a["Hour"] = a["RENT_DT"].dt.hour

  # "Rent Bike Count" Ïª¨Îüº ÏÉùÏÑ±
  df_fin = a.groupby(["Date","Hour"]).agg({"Hour":"count"})
  df_fin.rename(columns={"Hour":"Rent Bike Count"}, inplace=True)
  return df_fin.reset_index()
```

```python
df = make_col(df)
df
```
