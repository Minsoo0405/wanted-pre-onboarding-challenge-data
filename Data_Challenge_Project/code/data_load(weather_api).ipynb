### 2. 기상청 api를 html 파일로 저장하기 - 108(서울), 112(인천), 119(수원)

```python
from urllib.request import urlopen

def download_weather_api(year,mm,stn):
  # 변수 정의
  domain = "https://apihub.kma.go.kr/api/typ01/url/kma_sfctm3.php?"
  stn_id = f"stn={stn}&"
  option = "disp=0&help=0&authKey="
  auth = "nshgm0U3TXCIYJtFN31wIA"

  # 마지막 날이 31일인 달
  if int(mm) in [1,3,5,7,8,10,12]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}152300&"
    tm2 = f"tm1={year}{mm}160000&tm2={year}{mm}312300&"

    for i in [tm, tm2]:
      if i == tm:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # 파일로 저장
        file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")
      elif i == tm2:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # 파일로 저장
        file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}(2).html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} 폴더 내에 '{year}{mm}(2).html' 파일 저장 완료!!")

  # 마지막 날이 30일인 달
  elif int(mm) in [4,6,9,11]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}302300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")

  # 윤년이면서 2월인 경우
  elif (int(year) in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}292300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")

  # 윤년이 아니면서 2월인 경우
  elif (int(year) not in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}282300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")
```

```python
year_list = ["2018", "2019", "2020", "2021", "2022", "2023"]
mm_list = ["01", "02", "03", "04", "05", "06",
           "07", "08", "09", "10" ,"11" ,"12"]
stn_list = ["112","119"]

for i in year_list:
  for j in mm_list:
    for k in stn_list:
      download_weather_api(i, j, k)
```

### 3. html 파일을 csv 파일로 변환하기

```python
import pandas as pd

# 파일 경로 설정
file_path = "/content/drive/MyDrive/weather_data/2020/202002.html"

# 파일을 읽어와 데이터프레임으로 변환
df_1801 = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)

# 결과 출력
df_1801[1:-1]
```

```python
import os

def change_csv(year):

  # 기본 경로 설정
  base_path = f"/content/drive/MyDrive/weather_data/{year}/"

  # 빈 데이터프레임 생성
  all_data = pd.DataFrame()

  # 월별 파일 읽어오기
  for month in range(1, 13):

    if month in [1,3,5,7,8,10,12]:
      # 월 포맷팅 (1 -> 01)
      month_str = f"{month:02d}"

      # 파일 경로 설정
      file_path = os.path.join(base_path, f"{year}{month_str}.html")
      file_path2 = os.path.join(base_path, f"{year}{month_str}(2).html")

      # 파일을 읽어와 데이터프레임으로 변환
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)
      df_month2 = pd.read_csv(file_path2, encoding="cp949", delim_whitespace=True, skiprows=2)

      # 전체 데이터프레임에 추가
      all_data = pd.concat([all_data, df_month[1:-1], df_month2[1:-1]])
    else:
      # 월 포맷팅 (1 -> 01)
      month_str = f"{month:02d}"

      # 파일 경로 설정
      file_path = os.path.join(base_path, f"{year}{month_str}.html")

      # 파일을 읽어와 데이터프레임으로 변환
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)

      # 전체 데이터프레임에 추가
      all_data = pd.concat([all_data, df_month[1:-1]])

  # 결과 출력
  return all_data
```

```python
weather_df= pd.concat([change_csv(year) for year in year_list], ignore_index=True)
```

```python
# 지점번호(112,119) html 파일 데이터를 데이터프레임으로 가져오기
stn_list = ["112","119"]
year_list = ["2018", "2019", "2020", "2021", "2022", "2023"]

weather_112 = pd.concat([change_csv("112",year) for year in year_list], ignore_index=True)
weather_119 = pd.concat([change_csv("119",year) for year in year_list], ignore_index=True)
```

```python
def rename_columns(df, stn):
  new_columns = {f'{stn}_YYMMDDHHMI':f"{stn}_YYMMDDHHMI",
              f'{stn}_DAY':f"{stn}_Daily Rainfall(mm)",
              f'{stn}_DAY.1':f"{stn}_Minute of Daily Rainfall",
              f'{stn}_HR1':f"{stn}_Hourly Rainfall(mm)",
              f'{stn}_HR1.1':f"{stn}_Minute of Hourly Rainfall",
              f'{stn}_60M':f"{stn}_Accumulated Rainfall(mm) for max 60min",
              f'{stn}_60M.1':f"{stn}_Minute of Accumulated Rainfall"}

  new_df = df.rename(columns=new_columns)

  return new_df
```

```python
from urllib.request import urlopen

def download_api(year,mm,stn,var):
  # 변수 정의
  domain = "https://apihub.kma.go.kr/api/typ01/url/awsh.php?var=TA&"
  stn_id = f"stn={stn}&"
  option = "help=0&authKey="
  auth = "nshgm0U3TXCIYJtFN31wIA"

  # 마지막 날이 31일인 달
  if int(mm) in [1,3,5,7,8,10,12]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}152300&"
    tm2 = f"tm1={year}{mm}160000&tm2={year}{mm}312300&"

    for i in [tm, tm2]:
      if i == tm:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # 파일로 저장
        file_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/{year}{mm}.html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")
      elif i == tm2:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # 파일로 저장
        file_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/{year}{mm}(2).html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} 폴더 내에 '{year}{mm}(2).html' 파일 저장 완료!!")

  # 마지막 날이 30일인 달
  elif int(mm) in [4,6,9,11]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}302300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")

  # 윤년이면서 2월인 경우
  elif (int(year) in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}292300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")

  # 윤년이 아니면서 2월인 경우
  elif (int(year) not in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}282300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # 파일로 저장
    file_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} 폴더 내에 '{year}{mm}.html' 파일 저장 완료!!")
```

```python
import os

def change_csv(stn,year,var):

  # 기본 경로 설정
  base_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/"

  # 빈 데이터프레임 생성
  all_data = pd.DataFrame()

  # 월별 파일 읽어오기
  for month in range(1, 13):

    if month in [1,3,5,7,8,10,12]:
      # 월 포맷팅 (1 -> 01)
      month_str = f"{month:02d}"

      # 파일 경로 설정
      file_path = os.path.join(base_path, f"{year}{month_str}.html")
      file_path2 = os.path.join(base_path, f"{year}{month_str}(2).html")

      # 파일을 읽어와 데이터프레임으로 변환
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)
      df_month2 = pd.read_csv(file_path2, encoding="cp949", delim_whitespace=True, skiprows=2)

      # 전체 데이터프레임에 추가
      all_data = pd.concat([all_data, df_month[1:-1], df_month2[1:-1]])
    else:
      # 월 포맷팅 (1 -> 01)
      month_str = f"{month:02d}"

      # 파일 경로 설정
      file_path = os.path.join(base_path, f"{year}{month_str}.html")

      # 파일을 읽어와 데이터프레임으로 변환
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)

      # 전체 데이터프레임에 추가
      all_data = pd.concat([all_data, df_month[1:-1]])

  # 결과 출력
  return all_data
```
