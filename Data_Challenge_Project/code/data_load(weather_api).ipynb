```python
from urllib.request import urlopen

def download_weather_api(year,mm,stn):
  # ë³€ìˆ˜ ì •ì˜
  domain = "https://apihub.kma.go.kr/api/typ01/url/kma_sfctm3.php?"
  stn_id = f"stn={stn}&"
  option = "disp=0&help=0&authKey="
  auth = "nshgm0U3TXCIYJtFN31wIA"

  # ë§ˆì§€ë§‰ ë‚ ì´ 31ì¼ì¸ ë‹¬
  if int(mm) in [1,3,5,7,8,10,12]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}152300&"
    tm2 = f"tm1={year}{mm}160000&tm2={year}{mm}312300&"

    for i in [tm, tm2]:
      if i == tm:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # íŒŒì¼ë¡œ ì €ì¥
        file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} í´ë” ë‚´ì— '{year}{mm}.html' íŒŒì¼ ì €ì¥ ì™„ë£Œ!!")
      elif i == tm2:
        base_path = domain + i + stn_id + option + auth

        with urlopen(base_path) as f:
          content = f.read()

        # íŒŒì¼ë¡œ ì €ì¥
        file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}(2).html"

        with open(file_path, "wb") as file:
            file.write(content)
            print(f"{stn} í´ë” ë‚´ì— '{year}{mm}(2).html' íŒŒì¼ ì €ì¥ ì™„ë£Œ!!")

  # ë§ˆì§€ë§‰ ë‚ ì´ 30ì¼ì¸ ë‹¬
  elif int(mm) in [4,6,9,11]:
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}302300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # íŒŒì¼ë¡œ ì €ì¥
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} í´ë” ë‚´ì— '{year}{mm}.html' íŒŒì¼ ì €ì¥ ì™„ë£Œ!!")

  # ìœ¤ë…„ì´ë©´ì„œ 2ì›”ì¸ ê²½ìš°
  elif (int(year) in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}292300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # íŒŒì¼ë¡œ ì €ì¥
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} í´ë” ë‚´ì— '{year}{mm}.html' íŒŒì¼ ì €ì¥ ì™„ë£Œ!!")

  # ìœ¤ë…„ì´ ì•„ë‹ˆë©´ì„œ 2ì›”ì¸ ê²½ìš°
  elif (int(year) not in [2020]) & (int(mm) in [2]):
    tm = f"tm1={year}{mm}010000&tm2={year}{mm}282300&"

    base_path = domain + tm + stn_id + option + auth

    with urlopen(base_path) as f:
      content = f.read()

    # íŒŒì¼ë¡œ ì €ì¥
    file_path = f"/content/drive/MyDrive/weather_data/html_files/{stn}/{year}{mm}.html"

    with open(file_path, "wb") as file:
        file.write(content)
        print(f"{stn} í´ë” ë‚´ì— '{year}{mm}.html' íŒŒì¼ ì €ì¥ ì™„ë£Œ!!")
```

```python
import os

def change_csv(year):

  # ê¸°ë³¸ ê²½ë¡œ ì„¤ì •
  base_path = f"/content/drive/MyDrive/weather_data/{year}/"

  # ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
  all_data = pd.DataFrame()

  # ì›”ë³„ íŒŒì¼ ì½ì–´ì˜¤ê¸°
  for month in range(1, 13):

    if month in [1,3,5,7,8,10,12]:
      # ì›” í¬ë§·íŒ… (1 -> 01)
      month_str = f"{month:02d}"

      # íŒŒì¼ ê²½ë¡œ ì„¤ì •
      file_path = os.path.join(base_path, f"{year}{month_str}.html")
      file_path2 = os.path.join(base_path, f"{year}{month_str}(2).html")

      # íŒŒì¼ì„ ì½ì–´ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)
      df_month2 = pd.read_csv(file_path2, encoding="cp949", delim_whitespace=True, skiprows=2)

      # ì „ì²´ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
      all_data = pd.concat([all_data, df_month[1:-1], df_month2[1:-1]])
    else:
      # ì›” í¬ë§·íŒ… (1 -> 01)
      month_str = f"{month:02d}"

      # íŒŒì¼ ê²½ë¡œ ì„¤ì •
      file_path = os.path.join(base_path, f"{year}{month_str}.html")

      # íŒŒì¼ì„ ì½ì–´ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)

      # ì „ì²´ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
      all_data = pd.concat([all_data, df_month[1:-1]])

  # ê²°ê³¼ ì¶œë ¥
  return all_data
```

```python
def rename_columns(df, stn):
  new_columns = {f'{stn}_YYMMDDHHMI':f"{stn}_YYMMDDHHMI",
              f'{stn}_DAY':f"{stn}_Daily Rainfall(mm)",
              f'{stn}_DAY.1':f"{stn}_Minute of Daily Rainfall",
              f'{stn}_HR1':f"{stn}_Hourly Rainfall(mm)",
              f'{stn}_HR1.1':f"{stn}_Minute of Hourly Rainfall",
              f'{stn}_60M':f"{stn}_Accumulated Rainfall(mm) for max 60min",
              f'{stn}_60M.1':f"{stn}_Minute of Accumulated Rainfall"}

  new_df = df.rename(columns=new_columns)

  return new_df
```

```python
import os

def change_csv(stn,year,var):

  # ê¸°ë³¸ ê²½ë¡œ ì„¤ì •
  base_path = f"/content/drive/MyDrive/{var}_data/html_files/{stn}/"

  # ë¹ˆ ë°ì´í„°í”„ë ˆì„ ìƒì„±
  all_data = pd.DataFrame()

  # ì›”ë³„ íŒŒì¼ ì½ì–´ì˜¤ê¸°
  for month in range(1, 13):

    if month in [1,3,5,7,8,10,12]:
      # ì›” í¬ë§·íŒ… (1 -> 01)
      month_str = f"{month:02d}"

      # íŒŒì¼ ê²½ë¡œ ì„¤ì •
      file_path = os.path.join(base_path, f"{year}{month_str}.html")
      file_path2 = os.path.join(base_path, f"{year}{month_str}(2).html")

      # íŒŒì¼ì„ ì½ì–´ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)
      df_month2 = pd.read_csv(file_path2, encoding="cp949", delim_whitespace=True, skiprows=2)

      # ì „ì²´ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
      all_data = pd.concat([all_data, df_month[1:-1], df_month2[1:-1]])
    else:
      # ì›” í¬ë§·íŒ… (1 -> 01)
      month_str = f"{month:02d}"

      # íŒŒì¼ ê²½ë¡œ ì„¤ì •
      file_path = os.path.join(base_path, f"{year}{month_str}.html")

      # íŒŒì¼ì„ ì½ì–´ì™€ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
      df_month = pd.read_csv(file_path, encoding="cp949", delim_whitespace=True, skiprows=2)

      # ì „ì²´ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
      all_data = pd.concat([all_data, df_month[1:-1]])

  # ê²°ê³¼ ì¶œë ¥
  return all_data
```
