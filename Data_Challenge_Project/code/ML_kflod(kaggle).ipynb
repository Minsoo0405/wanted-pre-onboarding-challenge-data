```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import sklearn
```

# 1.데이터 준비

### 1.1 데이터 불러오기

```python
df = pd.read_csv('../data/SeoulBikeData_labeling_fin.csv', encoding='cp949')
df
```

### 1.2 날짜 데이터 형식 변환하기

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

### 1.3 범주형 데이터를 더미변수로 변경하기

```python
df = pd.get_dummies(df, columns=['Seasons', 'Holiday', 'Functioning Day'])
df
```

# 2.데이터 분할

### 2.1 데이터 분할하기

```python
X = df.drop(['Date', 'Year', 'Month', 'Day', 'Rented Bike Count'], axis=1)
y = df['Rented Bike Count']
```

```python
X = np.array(X)
y = np.array(y)
X, y
```

# 3. 교차검증

### 3.1 라이브러리 임포트

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold

from sklearn.metrics import r2_score

from sklearn.model_selection import train_test_split
```

### 3.2 모델 

```python
lr = LinearRegression()
```

```python
kfold = KFold(5)
list(kfold.split(X))
```

```python
r2_scores = []

for train_idx, test_idx in kfold.split(X):
    X_train, X_test = X[train_idx], X[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    reg = lr.fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    y_pred[y_pred < 0] = 0.
    
    r2 = r2_score(y_test, y_pred)
    r2_scores.append(r2)
```

```python
r2_scores
```

```python
for i, r2 in enumerate(r2_scores):
    print(i+1, f'- R2 = {r2:.3f}')
    
print(f'average R2 = {np.round(np.mean(r2_scores),3)}')
```
